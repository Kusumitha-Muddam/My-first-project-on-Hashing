**5.FEATURE ENGINEERING(DATE->YEAR/MONTH/DAY OF WEEK)**

```python
#Convert Date to Datetime and extracyt uuseful componenets
x['date']=pd.to_datetime(x['date'],errors="coerce",dayfirst=True)
x['soldyear']=x['date'].dt.year
x['soldmonth']=x['date'].dt.month
x['soldday']=x['date'].dt.day

# Drop row date afetr extracting features
x.drop(columns=['date'],inplace=True)

```

**6.SELECT NUMERIC VS CATEGORICAL COLUMNS**

```python
num_x=x.select_dtypes(include="number")
cat_x=x.select_dtypes(exclude="number")
print(cat_x)
```

**7 PREPROCESSING PIPELINE(IMPUTE+ENCODE+SCALE)**

```python
# Numeric:median impute+scale
num_transformation=Pipeline(steps=[("imputer",SimpleImputer(strategy="median")),("scaling",StandardScaler())])
# Categorical:most_frequent(mode) impute+one-hot encode
cat_transformation=Pipeline(steps=[("imputer",SimpleImputer(strategy="most_frequent")),("encoding",OneHotEncoder(handle_unknown="ignore"))])
preprocessor=ColumnTransformer(transformers=[("num",num_transformation,num_x),("cat",cat_transformation,cat_x)],remainder="drop")
preprocessor
```
